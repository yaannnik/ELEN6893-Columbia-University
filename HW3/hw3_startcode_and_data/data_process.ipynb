```python
node_file = "nodes.csv"
edge_file = "edges.csv"
```

```python
if os.path.isfile(node_file):
    os.remove(node_file)

if os.path.isfile(edge_file):
    os.remove(edge_file)

with open("nodes.csv", "a") as f:
    csv_writer = csv.writer(f)
    csv_writer.writerow(["node"])

with open("edges.csv", "a") as f:
    csv_writer = csv.writer(f)
    csv_writer.writerow(["source", "target"])
```

```python
nodes = {}
```

```python
order = 0
for _ in data:
    p1 = _[0]
    p2 = _[1]

    if p1 not in nodes.keys():
        nodes[p1] = order
        with open("nodes.csv", "a") as f:
            csv_writer = csv.writer(f)
            csv_writer.writerow([p1])
        order += 1

    if p2 not in nodes.keys():
        nodes[p2] = order
        with open("nodes.csv", "a") as f:
            csv_writer = csv.writer(f)
            csv_writer.writerow([p2])
        order += 1

    with open("edges.csv", "a") as f:
        csv_writer = csv.writer(f)
        csv_writer.writerow([nodes[p1], nodes[p2]])
```
